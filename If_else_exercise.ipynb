```python
"""Q1. Write a program to calculate the electricity bill 
using only if statement? (accept number of unit from user) 
according to the following criteria:
 Unit Price 
First 100 units no charge
Next 100 units Rs 5 per unit
After 200 units Rs 10 per unit
(For example, if input unit is 350 than total bill amount is Rs2000). Lets explain…
For first 100 no charges
Now you have 250 unit remaining
100-200 you need to pay – 100*5 = 500
Now 200-100 you need to pay 150*10 = 1500
Total = 500+1500 = 2000"""

No_of_Units = int(input("Enter the no. of units : "))
if No_of_Units <= 100:
    print("No charge")
elif No_of_Units > 100 and No_of_Units <= 200:
    print("Bill due : ",(No_of_Units - 100)* 5 )
elif No_of_Units > 200:
    print("Bill due : ",(No_of_Units - 200) * 10 + (100 * 5))
```

```python
"Using only if statement"
Units = int(input("Enter Units : "))
if Units >200:
    print("Bill Due :",(Units-200) *10 + (100 *5))
if Units > 100 and Units <= 200:
    print("Bill Due :",(Units - 100)*5)
if Units <= 100:
    print("No Charge")
```

```python
"""Q2. Write a program to accept percentage from the user and display 
the grade according to the following criteria:
 Marks Grade
 > 90 A
 > 80 and <= 90 B
 >= 60 and <= 80 C
 below 60 D"""

Marks = int(input("Enter Marks : "))
if Marks > 90:
    print("Grade A")
elif Marks >80 and Marks <= 90:
    print("Grade B")
elif Marks >= 60 and Marks <= 80:
    print("Grade C")
else:
    print("Grade D")
```

```python
"""Q4. A company decided to give bonus to employee 
according to following criteria:
 Time period of Service      Bonus

 More than 10 years           10%
 >=6 and <=10                  8%
 Less than 6 years             5%

Ask user for their salary and years of service and print the 
net bonus amount?"""

T_P_S = int(input("Enter the Time Period Of Service : "))
Salary = int(input("Enter Your Salary : "))

if T_P_S > 10:
    print("Net Bonus Amount : ", Salary * 0.10)
elif T_P_S >=6 and T_P_S <= 10:
    print("Net Bonus Amount : ", Salary * 0.08)
else:
    print("Net Bonus Amount : ", Salary * 0.05)
```

```python
"""Q5. Accept three numbers from the user and 
display the second largest number?"""

number_1 = int(input("1st Input: "))
number_2 = int(input("2nd Input: "))
number_3 = int(input("3rd Input: "))

if (number_1 == number_2) and (number_2 == number_3):
    print("All numbers are equal")

elif (number_1 > number_2 and number_1 < number_3) or (number_1 < number_2 and number_1 > number_3):
    print(number_1 , "is the second largest number.")
elif (number_2 > number_1 and number_2 < number_3) or (number_2 < number_1 and number_2 > number_3):
    print(number_2 , "is the second largest number.")
else:
    print(number_3 , "is the second largest number.")
```

```python
 """Q7. Accept the marks of English, Math and Science, 
Social Studies Subject and display the stream allotted 
according to following:

All Subjects more than 80 marks — Science Stream
English >80 and Math, Science above 50 — Commerce Stream
English > 80 and social studies > 80 — Humanities"""


english = int(input("Enter Eng marks : "))
maths = int(input("Enter Maths marks : "))
science = int(input("Enter Sci marks : "))
social_studies = int(input("Enter SST marks : "))

if english > 80 and maths > 80 and science > 80 and social_studies >80:
    print("Science Stream")
elif english > 80 and maths > 50 and science > 50:
    print("Commerce Stream")
elif english > 80 and social_studies > 80:
    print("Humanities")
else:
    print("Didn't match the requirements")
```
