```python
"""1. Electricity Bill (variant)

Write a program to calculate the electricity bill:

First 50 units – ₹0.50 per unit

Next 100 units – ₹0.75 per unit

Next 100 units – ₹1.20 per unit

Above 250 units – ₹1.50 per unit
Add a fixed charge of ₹50 to the bill at the end."""

fixed_charge = 50
unit = 500

if unit <= 50 :
    print("Bill Due : ",(unit * 0.50)+ fixed_charge)

elif unit <= 150 :
    print("Bill Due : ",(50 * 0.50)+ (unit-50)*0.75 + fixed_charge)

elif unit <= 250:
    print("Bill Due : ",(50 * 0.50)+(100*0.75) + (unit-150)*1.20 +fixed_charge)

else:
    print("Bill Due : ",(100*1.20) +(100*0.75) + (50 * 0.50)+(unit-250)*1.50 + fixed_charge)
```

```python
"""2. Income Tax Calculator

Ask the user for their annual income and calculate tax as follows:

Up to ₹2,50,000 → No tax

₹2,50,001 – ₹5,00,000 → 5%

₹5,00,001 – ₹10,00,000 → 20%

Above ₹10,00,000 → 30%"""


annual_income= 1200000
if annual_income <= 250000:
    print("no tax")

elif annual_income <= 500000:
    print("Tax : ",annual_income *5//100)

elif annual_income <= 1000000:
    print("Tax : ",annual_income *20//100)

else:
    print("Tax : ",annual_income*30//100)
```
